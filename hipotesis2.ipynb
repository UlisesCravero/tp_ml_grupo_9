```python
try:
    import pandas as pd
    import numpy as np
    import matplotlib.pyplot as plt
    import seaborn as sns
    print("las dependencias se han instalado correctamente")
except Exception as e:
    print(f"Error al importar las dependencias: {e}")
```

```python
dfs = [pd.read_csv(f'parte_{i}.csv') for i in range(1, 6)]

# Unir todas las partes en un solo DataFrame y guardarlo en la variable 'frauds'
frauds = pd.concat(dfs, ignore_index=True)
frauds.sample(5)
```

```python
frauds_sin_todos_balances_cero = frauds[
                        (frauds['oldbalanceOrg'] != 0) &
                        (frauds['newbalanceOrig'] != 0) &
                        (frauds['oldbalanceDest'] != 0) &
                        (frauds['oldbalanceDest'] != 0)]
```

```python
cash_out = frauds_sin_todos_balances_cero[(frauds_sin_todos_balances_cero['type'] == 'CASH_OUT')]

print("cantidad en dataset: ", cash_out.shape)

cash_out.sample(20)
```

```python
frauds_sin_todos_balances_cero_doscolumnas = cash_out[
                        ((cash_out['oldbalanceOrg'] == 0) &
                        (cash_out['newbalanceOrig'] == 0)) |
                        ((cash_out['oldbalanceDest'] == 0) &
                        (cash_out['newbalanceDest'] == 0))]

                        
print("cantidad en dataset frauds_sin_todos_balances_cero: ", frauds_sin_todos_balances_cero_doscolumnas.shape)

```

```python
frauds_sin_todos_balances_cero_AlgunaColumnas = cash_out[
                        (cash_out['oldbalanceOrg'] == 0) |
                        (cash_out['newbalanceOrig'] == 0) |
                        (cash_out['oldbalanceDest'] == 0) |
                        (cash_out['newbalanceDest'] == 0)]

                                                
print("cantidad en dataset frauds_sin_todos_balances_cero_AlgunaColumnas: ", frauds_sin_todos_balances_cero_AlgunaColumnas.shape)

frauds_sin_todos_balances_cero_AlgunaColumnas.sample(11)
```
